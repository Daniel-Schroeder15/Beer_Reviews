```python
# Remove unnecessary columns
beer_data = beer_data.drop(["brewery_id"], axis=1)
beer_data = beer_data.drop(["review_time"], axis=1)
beer_data = beer_data.drop(["review_profilename"], axis=1)
beer_data = beer_data.drop(["beer_beerid"], axis=1)
beer_data = beer_data.dropna()
beer_data.head()
```

```python
beer_data.info()
```

```python
beer_data.describe()
```

```python
#Rearrange and sort by brewery name

beer_data = beer_data[['brewery_name', 'beer_style', 'beer_name', 'beer_abv', 'review_appearance', 
             'review_aroma', 'review_palate', 'review_taste', 'review_overall']]
beer_data = beer_data.sort_values(by=['brewery_name', 'beer_style', 'beer_name' , 'beer_abv', 'review_overall'])
beer_data = beer_data.reset_index()
beer_data = beer_data.drop(["index"], axis=1)
beer_data.head()
```

```python
beer_data.tail()
```

```python
# created new column.. review average
beer_data['review_average'] = beer_data.apply(lambda row: (row["review_overall"] + row["review_aroma"] + 
                                                 row["review_appearance"] + row["review_palate"] + 
                                                 row["review_taste"]) / 5, axis=1)

beer_data = beer_data.drop(beer_data[(beer_data["review_average"] < 1) | (beer_data["review_average"] > 5)].index)
beer_data.head()
```
